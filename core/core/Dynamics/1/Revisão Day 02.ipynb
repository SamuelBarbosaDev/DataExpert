## Exemplos: Prática

```python
# Função para mostrar métodos e atributos de um objeto
def mostrar_metodos_e_atributos(obj):
  """Recebe objeto e retorna um dicionário listando os métodos e atributos do objeto"""
  metodos_e_atributos = {'Métodos': [], 'Atributos': []}
  # Listar atributos ignorando métodos específicos do Python
  atributos = [attr for attr in dir(obj) if not attr.startswith('_')]
  # Listar métodos e atributos
  for attr in atributos:
    if callable(getattr(obj, attr)):
      metodos_e_atributos['Métodos'].append(attr)
    else:
      metodos_e_atributos['Atributos'].append(attr)
  return metodos_e_atributos
```

```python
clientes = ['Alan', 'Jorge', 'Maria', 'João', 'Talita']
gastos = [45, 30, 22.5, 18, 55.25, 50]
```

```python
mostrar_metodos_e_atributos(clientes)
```

Qual foi a média de gastos?

```python
sum(gastos) / len(gastos)
```

Qual cliente gastou mais, e quanto gastou?

```python
menor_gasto = min(gastos)
maior_gasto = max(gastos)

menor_gasto_idx = gastos.index(menor_gasto)
maior_gasto_idx = gastos.index(maior_gasto)

print(f'''
O cliente com menor gasto foi {clientes[menor_gasto_idx]} (R$ {menor_gasto:.2f}).
O cliente com maior gasto foi {clientes[maior_gasto_idx]} (R$ {maior_gasto:.2f}).
''')
```
