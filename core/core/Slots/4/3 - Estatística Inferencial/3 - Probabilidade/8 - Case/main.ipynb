```python
import random
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as st
from scipy.stats import norm
```

```python
class  EstatísticaInferencial():
    def __init__(self):
        path = r'../../data/n_movies.csv'
        self.df = pd.read_csv(
            filepath_or_buffer=path,
            sep=',',
            decimal='.'
        )

        self.df = self.df.dropna()

    def dataframe(self):
        display(
            self.df.head(5)
        )

    def descritiva(self):
        df = self.df.describe()
        display(
            df
        )
    
    def infos(self):
        df = self.df.info()
        display(
            df
        )
    
    def hisplot_rantig(self):
        sns.histplot(
            data=self.df,
            x='rating',
            kde=True
        )

    def boxplot_rantig(self):
        sns.boxplot(
            data=self.df,
            x='rating'
        )

    def tamanho_da_amostra(self):
        z = 1.96
        p = 0.5
        e = 0.05
        n = 5754

        tamanho_da_amostra = (
            (z**2)*p*(1-p)/(e**2))/1+(((z**2)*p*(1-p))/((e**2)*n)
        )

        print(tamanho_da_amostra)
    
    def amostra_simples(self):
        amostra = self.df.sample(385, replace=True)

        display(
            amostra
        )
    
    def amostra_sistemática(self):
        indexes = np.arange(
            start=5, 
            stop=len(self.df), 
            step=5,
        )

        amostra_sistemática = self.df.iloc[indexes]
        display(
            amostra_sistemática
        )
    
    def teorema_do_limite_central(self):
        lista = self.df['rating'].values.tolist()
        means = []

        for i in range(20):
            amostra = random.sample(
                population=lista, 
                k=30
            )
            means.append(np.mean(amostra)) 

        sns.histplot(
            data=means,
            kde = True,
        )

    def intervalo_de_confiaça(self):
        lista = self.df['rating'].values.tolist()
        means = []

        for i in range(20):
            amostra = random.sample(
                population=lista, 
                k=30
            )
            means.append(np.mean(amostra))

        teste_de_normalização_95 = st.norm.interval(
            alpha=0.95,
            loc=np.mean(means),
            scale=st.sem(means),
        )

        print(f'intervalo_de_confiança:{teste_de_normalização_95}')

if __name__ == "__main__":
    estatística_inferencial = EstatísticaInferencial()
```
