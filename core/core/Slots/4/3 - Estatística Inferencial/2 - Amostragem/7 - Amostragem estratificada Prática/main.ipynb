```python
import numpy as np
import pandas as pd
```

```python
class  EstatísticaInferencial():
    def __init__(self):
        path = r'../../data/Iris.csv'
        self.df = pd.read_csv(
            filepath_or_buffer=path,
            sep=',',
            decimal='.',
        )

    def dataframe(self):
        display(
            self.df.head(10)
        )
    
    def amostra_aleatória(self):
        amostra = self.df.sample(
            n=30,
            replace=True,
        )

        display(
            amostra.head(10)
        )

    def amostra_sistemática_aleatória(self):
        amostra = self.df.sample(
            n=30,
            replace=True,
        )
        
        indexes = np.arange(
            start=3, 
            stop=len(amostra), 
            step=3,
        )

        amostra_sistemática = amostra.iloc[indexes]
        display(
            amostra_sistemática
        )
    
    def amostra_estratificada_aleatória(self):
        amostra = self.df.groupby(
            by='Species',
            group_keys=False,
        ).apply(lambda x : x.sample(2))

        display(
            amostra
        )


if __name__ == "__main__":
    estatística_inferencial = EstatísticaInferencial()
```

```python
estatística_inferencial.amostra_estratificada_aleatória()
```
