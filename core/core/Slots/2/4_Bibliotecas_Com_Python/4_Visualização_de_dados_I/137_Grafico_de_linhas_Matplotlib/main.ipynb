```python
import pandas as pd
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
class GraficoDeLinhasMatplotlib():
    """
        Grafico de linhas - Matplotlib
    """
    def __init__(self):
        self.path = r'../DataFrame/clientes_rename.xlsx'
        self.credit = pd.read_excel(
            self.path
        )
        self.credit_group = self.credit.groupby(
            ["idade"]
        ).mean().reset_index()

    def grafico_com_metplotlib(self):
        plt.figure(
            figsize=(20, 15)
        )
        plt.xlabel('Idade')
        plt.ylabel('Valor Do Credito')
        plt.plot(
            self.credit_group['idade'],
            self.credit_group['valor_credito'],  
        )
        plt.grid(
            b=True, 
            color='gray',
            linestyle='--',
            linewidth=1
        )
```

```python
if __name__ == '__main__':
    grafico_de_linhas_matplotlib = GraficoDeLinhasMatplotlib()
    grafico_de_linhas_matplotlib.grafico_com_metplotlib()
```
