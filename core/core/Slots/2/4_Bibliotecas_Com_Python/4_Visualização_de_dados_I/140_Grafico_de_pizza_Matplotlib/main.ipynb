```python
import pandas as pd
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
class GraficoDePizzaMatplotlib():
    """
        Grafico de pizza - Matplotlib

    """
    def __init__(self):
        self.path = r'../DataFrame/clientes_rename.xlsx'
        self.credit = pd.read_excel(
            self.path
        )
        self.credit_group = self.credit.groupby(
            ["sexo"]
        ).mean().reset_index()
    
    def grafico_com_metplotlib(self):
        fig, ax = plt.subplots(
            figsize=(10, 10),
        )
        ax.pie(
            self.credit_group['valor_credito'],
            labels=self.credit_group['sexo'],
            autopct='%.2f%%',
            shadow=True,
            startangle=0,
            colors=['blue', 'red']
        )
```

```python
if __name__ == '__main__':
    grafico_de_pizza_matplotlib = GraficoDePizzaMatplotlib()
    grafico_de_pizza_matplotlib.grafico_com_metplotlib()
```
