```python
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
class GráficoDeViolin():
    """
        Gráfico de violin
    """
    def __init__(self):
        self.path = r'../DataFrame/clientes_rename.xlsx'
        self.credit = pd.read_excel(
            self.path
        )
        self.credit_group = self.credit.groupby(
            ["sexo"]
        ).mean().reset_index()
    
    def grafico_com_plotly_violin(self):
        pass

    def analise_de_credito_por_estado_civil(self):
        fig = px.violin(
            self.credit,
            x='estado_civil',
            y='valor_credito',
            color='sexo',
            box=True,
            title='Credito Por Estado Civil'
        )
        fig.show()
```
