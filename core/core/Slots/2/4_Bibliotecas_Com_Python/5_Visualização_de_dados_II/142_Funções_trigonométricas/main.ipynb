```python
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
class FunçõesTrigonométricas():
    """
        Funções trigonométricas
    """
    def __init__(self):
        self.path = r'../DataFrame/clientes_rename.xlsx'
        self.credit = pd.read_excel(
            self.path
        )
        self.credit_group = self.credit.groupby(
            ["sexo"]
        ).mean().reset_index()
    
    def grafico_com_matplotlib(self):
        matriz = np.linspace(0.0, 2.0, 100)
        matriz_pi = np.sin(np.pi * matriz)
        matriz_vezes_pi = np.sin(2 * np.pi * matriz)

        plt.figure(
            figsize=(20, 8)
        )
        plt.plot(
            matriz,
            matriz_pi,
            linewidth=2.0,
            label=r"$f_1(matriz)=sin(\pi \times matriz)$"
        )
        plt.plot(
            matriz,
            matriz_vezes_pi,
            linewidth=2.0,
            label=r"$f_2(matriz)=sin(2 \times \pi \times matriz)$"
        )

        plt.legend()
        
        plt.grid(
            b=True, 
            color='gray', 
            linestyle='--', 
            linewidth=0.5,
        )
```

```python
if __name__ == '__main__':
    funções_trigonométricas = FunçõesTrigonométricas()
    funções_trigonométricas.grafico_com_matplotlib()
```
