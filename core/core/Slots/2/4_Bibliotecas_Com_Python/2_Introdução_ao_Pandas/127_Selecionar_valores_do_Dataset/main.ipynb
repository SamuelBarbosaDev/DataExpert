```python
import pandas as pd
```

```python
class SelecionarValoresDoDataset():
    """
        Selecionar Valores Do Dataset
    """
    def __init__(self):
        self.path = r'../Dataset_colabs/clientes_rename.xlsx'
        self.credit = pd.read_excel(
            self.path
        )
    
    def acessando_uma_coluna(self):
        print(f'\nAcessando Coluna: {self.credit["valor_credito"]}')
    
    def acessando_uma_coluna_de_outra_forma(self):
        print(f'\nAcessando Coluna: {self.credit}')

    def acessando_varias_colunas(self):
        print(f'\nAcessando Coluna:\n {self.credit[["valor_credito", "nivel_escolaridade"]]}')
    
    def filtrando_colunas(self):
        filtro = self.credit.filter(
            like='pag'
        )
        print(f'\nAcessando Coluna: {filtro}')

```

```python
if __name__ == '__main__':
    selecionar_valores_do_dataset = SelecionarValoresDoDataset()
    selecionar_valores_do_dataset.acessando_uma_coluna()
    selecionar_valores_do_dataset.acessando_uma_coluna_de_outra_forma()
    selecionar_valores_do_dataset.acessando_varias_colunas()
    selecionar_valores_do_dataset.filtrando_colunas()
```
