```python
import pandas as pd
```

```python
class AnaliseDeDadosComPandasI():
    """
        Analise De Dados Com Pandas I
    """
    def __init__(self):
        self.path = r'../DataFrame/credit_card_clients_2.csv'
        self.credit = pd.read_csv(
            self.path
        )

    def retorna_com_true_os_valores_vazios(self):
        print(self.credit.isnull())

    def retorna_com_true_os_valores_não_vazios(self):
        print(self.credit.notnull())

    def somando_os_valores_nulos(self):
        print(self.credit.isnull().sum())
    
    def filtrando_os_valores_nulos(self):
        print(self.credit[self.credit['ID'].isnull()])

    def excluindo_as_linhas_nulas(self):
        credit = self.credit
        print(credit[credit['ID'].notna()])
    
    def excluindo_valores_nulos(self):
        credit = self.credit
        print(credit.dropna(how='all'))
        

```

```python
if __name__ == '__main__':
    analise_de_dados_com_pandas_i = AnaliseDeDadosComPandasI()
    analise_de_dados_com_pandas_i.retorna_com_true_os_valores_vazios()
    analise_de_dados_com_pandas_i.retorna_com_true_os_valores_não_vazios()
    analise_de_dados_com_pandas_i.somando_os_valores_nulos()
    analise_de_dados_com_pandas_i.filtrando_os_valores_nulos()
    analise_de_dados_com_pandas_i.excluindo_as_linhas_nulas()
    analise_de_dados_com_pandas_i.excluindo_valores_nulos()
```
