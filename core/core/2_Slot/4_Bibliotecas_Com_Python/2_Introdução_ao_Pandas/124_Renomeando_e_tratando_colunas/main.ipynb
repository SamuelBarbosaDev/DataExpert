```python
import pandas as pd
```

```python
class RenomeandoETratandoColunas():
    """
        Extraindo Informações Do Dataset.
    """
    def __init__(self):
        self.path = r'../Dataset_colabs/credit_card_clients copy.csv'
        self.credit = pd.read_csv(
            self.path
        )
    def renomeando_dataset(self):
        self.credit_rename = self.credit.rename(
            columns= {
                'LIMIT_BAL':'valor_credito',
                'SEX':'sexo',
                'EDUCATION':'nivel_escolaridade',
                'MARRIAGE':'estado_civil',
                'AGE':'idade',
                'PAY_0':'status_pag_set_2005',
                'PAY_2':'status_pag_ago_2005',
                'PAY_3':'status_pag_jul_2005',
                'PAY_4':'status_pag_jun_2005',
                'PAY_5':'status_pag_mai_2005',
                'PAY_6':'status_pag_abr_2005',
                'BILL_AMT1':'extrato_set_2005',
                'BILL_AMT2':'extrato_ago_2005',
                'BILL_AMT3':'extrato_jul_2005',
                'BILL_AMT4':'extrato_jun_2005',
                'BILL_AMT5':'extrato_mai_2005',
                'BILL_AMT6':'extrato_abr_2005',
                'PAY_AMT1':'valor_pag_set_2005',
                'PAY_AMT2':'valor_pag_ago_2005',
                'PAY_AMT3':'valor_pag_jul_2005',
                'PAY_AMT4':'valor_pag_jun_2005',
                'PAY_AMT5':'valor_pag_mai_2005',
                'PAY_AMT6':'valor_pag_abr_2005'
            }
        )

        print(self.credit_rename)

    def excluindo_coluna(self):
        self.credit_rename.drop(
            ['default payment next month'], 
            axis=1, 
            inplace=True
        )
    
    def salvando_no_formato_xlsx(self):
        path = r'../Dataset_colabs/clientes_rename.xlsx'
        self.credit_rename.to_excel(
            path
        )

    
```

```python
if __name__ == '__main__':
    renomeando_e_tratando_colunas = RenomeandoETratandoColunas()
    renomeando_e_tratando_colunas.renomeando_dataset()
    renomeando_e_tratando_colunas.excluindo_coluna()
    renomeando_e_tratando_colunas.salvando_no_formato_xlsx()
```
