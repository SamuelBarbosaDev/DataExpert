```python
import pandas as pd
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
class GraficoDeBarrasMatplotlib():
    """
        Grafico De Barras Matplotlib
    """
    def __init__(self):
        self.path = r'../DataFrame/clientes_rename.xlsx'
        self.credit = pd.read_excel(
            self.path
        )

    def agrupando_dados(self):
        self.credit_group = self.credit.groupby(["nivel_escolaridade"]).mean().reset_index()
        # print(f'\nAgrupand Dados:\n{self.credit_group}')
    
    def grafico_media_de_credito_por_escolaridade(self):
        plt.figure(
            figsize=(20, 10)
        )
        plt.title('Grafico De Barras Matplotlib :)')
        plt.xlabel("nivel_escolaridade")
        plt.ylabel("Média do valor de crédito concedido")
        plt.bar(
            self.credit_group["nivel_escolaridade"],
            self.credit_group['valor_credito'],
            color="rEd"
        )
```

```python
if __name__ == '__main__':
    grafico_de_barras_matplotlib = GraficoDeBarrasMatplotlib()
    grafico_de_barras_matplotlib.agrupando_dados()
    grafico_de_barras_matplotlib.grafico_media_de_credito_por_escolaridade()
```
