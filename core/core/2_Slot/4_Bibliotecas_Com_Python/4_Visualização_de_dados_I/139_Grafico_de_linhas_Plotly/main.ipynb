```python
import pandas as pd
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
class GraficoDeLinhasPlotly():
    """
        Grafico de linhas - Plotly
    """
    def __init__(self):
        self.path = r'../DataFrame/clientes_rename.xlsx'
        self.credit = pd.read_excel(
            self.path
        )
        self.credit_group = self.credit.groupby(
            ["idade"]
        ).mean().reset_index()
    
    def grafico_com_a_plotly(self):
        fig =  make_subplots(
            rows=3,
            cols=2,
            shared_yaxes=True,
            subplot_titles=(
                'Abril',
                'Maior',
                'Junho',
                'Julho',
                'Agosto',
                'Setembro',
            )
        )
        fig.add_trace(
            go.Scatter(
                x=self.credit_group['idade'],
                y=self.credit_group['valor_pag_abr_2005'],
                line=dict(
                    color='red',
                    width=1.5,
                ),
                name='Valor Pago 04/2005',
            ),           
            row=1,
            col=1,
        )

        fig.add_trace(
            go.Scatter(
                x=self.credit_group['idade'],
                y=self.credit_group['valor_pag_mai_2005'],
                line=dict(
                    color='blue',
                    width=1.5,
                ),
                name='Valor Pago 05/2005',
            ),           
            row=1,
            col=2,
        )

        fig.add_trace(
            go.Scatter(
                x=self.credit_group['idade'],
                y=self.credit_group['valor_pag_jun_2005'],
                line=dict(
                    color='green',
                    width=1.5,
                ),
                name='Valor Pago 06/2005',
            ),           
            row=2,
            col=1,
        )

        fig.add_trace(
            go.Scatter(
                x=self.credit_group['idade'],
                y=self.credit_group['valor_pag_jul_2005'],
                line=dict(
                    color='gray',
                    width=1.5,
                ),
                name='Valor Pago 07/2005',
            ),           
            row=2,
            col=2,
        )

        fig.add_trace(
            go.Scatter(
                x=self.credit_group['idade'],
                y=self.credit_group['valor_pag_ago_2005'],
                line=dict(
                    color='violet',
                    width=1.5,
                ),
                name='Valor Pago 08/2005',
            ),           
            row=3,
            col=1,
        )

        fig.add_trace(
            go.Scatter(
                x=self.credit_group['idade'],
                y=self.credit_group['valor_pag_set_2005'],
                line=dict(
                    color='sienna',
                    width=1.5,
                ),
                name='Valor Pago 09/2005',
            ),           
            row=3,
            col=2,
        )

        fig.update_layout(
            title_text='Valor Pago Em 2005 por idade e por mês'
        )
        fig.update_yaxes(
            title_text='$ Dolar $'
        )
        fig.show()
        fig.write_html('grafico_valor_pago.html')
```

```python
if __name__ == '__main__':
    grafico_de_linhas_plotly = GraficoDeLinhasPlotly()
    grafico_de_linhas_plotly.grafico_com_a_plotly()
    
```
