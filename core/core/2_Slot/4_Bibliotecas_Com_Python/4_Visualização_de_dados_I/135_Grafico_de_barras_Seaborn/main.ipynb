```python
import pandas as pd
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
class GraficoDeBarrasSeaborn():
    """
        Grafico De Barras Seaborn
    """
    def __init__(self):
        self.path = r'../DataFrame/clientes_rename.xlsx'
        self.credit = pd.read_excel(
            self.path
        )
        self.credit_group = self.credit.groupby(
            ["nivel_escolaridade"]
        ).mean().reset_index()

    def extrato_bancario_dos_clientes(self):
        fig, axes = plt.subplots(2, 2, figsize=(20, 15))
        sns.barplot(
            data=self.credit_group, 
            x='nivel_escolaridade',
            y='extrato_set_2005',
            ax=axes[0, 0]
        )

        sns.barplot(
            data=self.credit_group, 
            x='nivel_escolaridade',
            y='extrato_ago_2005',
            ax=axes[0, 1]
        )

        sns.barplot(
            data=self.credit_group, 
            x='nivel_escolaridade',
            y='extrato_jul_2005',
            ax=axes[1, 0]
        )

        sns.barplot(
            data=self.credit_group, 
            x='nivel_escolaridade',
            y='extrato_jun_2005',
            ax=axes[1, 1]
        )
```

```python
if __name__ == '__main__':
   grafico_de_barras_Seaborn = GraficoDeBarrasSeaborn()
   grafico_de_barras_Seaborn.extrato_bancario_dos_clientes()
```
