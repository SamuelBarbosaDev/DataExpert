```python
import pandas as pd
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
class GraficoDeLinhasSeaborn():
    """
        Grafico de linhas - Seaborn
    """
    def __init__(self):
        self.path = r'../DataFrame/clientes_rename.xlsx'
        self.credit = pd.read_excel(
            self.path
        )
        self.credit_group = self.credit.groupby(
            ["idade"]
        ).mean().reset_index()

    def grafico_com_seaborn(self):
        fig, axes = plt.subplots(
            2,
            2,
            figsize=(20, 10),
        )
        sns.lineplot(
            data=self.credit,
            x='idade',
            y='extrato_jun_2005',
            hue='sexo',
            style="sexo",
            palette='Set2',
            ax=axes[0, 0]
        ).set(title='Junho')
        
        sns.lineplot(
            data=self.credit,
            x='idade',
            y='extrato_jul_2005',
            hue='sexo',
            style="sexo",
            palette='Set2',
            ax=axes[0, 1]
        ).set(title='Julho')

        sns.lineplot(
            data=self.credit,
            x='idade',
            y='extrato_ago_2005',
            hue='sexo',
            style="sexo",
            palette='Set2',
            ax=axes[1, 0]
        ).set(title='Agosto')

        sns.lineplot(
            data=self.credit,
            x='idade',
            y='extrato_set_2005',
            hue='sexo',
            style="sexo",
            palette='Set2',
            ax=axes[1, 1]
        ).set(title='Setembro')
```

```python
if __name__ == '__main__':
    grafico_de_linhas_seaborn = GraficoDeLinhasSeaborn()
    grafico_de_linhas_seaborn.grafico_com_seaborn()
```
