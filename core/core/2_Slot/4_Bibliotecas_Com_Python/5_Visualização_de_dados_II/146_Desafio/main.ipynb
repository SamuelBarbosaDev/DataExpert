```python
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
class Desafio():
    """
        Desafio
    """
    def __init__(self):
        self.path = r'../DataFrame/clientes_rename.xlsx'
        self.credit = pd.read_excel(
            self.path
        )
        self.credit_group = self.credit.groupby(
            ["sexo"]
        ).mean().reset_index()

    def histograma(self):
        fig = plt.figure(
            figsize=(10, 5)
        )
        sns.histplot(
            data=self.credit,
            x='idade',
            kde=True,
            hue='sexo',
            palette='dark'
        )
        plt.legend(
            labels=[
                'Masculino',
                'Feminino'
            ]
        )

    def imprimir(self):
        print(self.credit)
```

```python
if __name__ == '__main__':
    desafio = Desafio()
    desafio.histograma()
```
