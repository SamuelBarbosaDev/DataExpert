```python
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
class DiagramaDeCorrelação():
    """
        Diagrama de correlação
    """
    def __init__(self):
        self.path = r'../DataFrame/clientes_rename.xlsx'
        self.credit = pd.read_excel(
            self.path
        )
        self.credit_group = self.credit.groupby(
            ["sexo"]
        ).mean().reset_index()

    def excluindo_colunas(self):
        self.credit.drop(['Unnamed: 0', 'ID'], axis=1, inplace=True)
        print(self.credit)

    
    def grafico_com_seaborn(self):
        plt.figure(
            figsize=(20, 20)
        )

        sns.heatmap(
            self.credit.corr(),
            cmap='coolwarm',
            annot=True,
        )
```

```python
if __name__ == '__main__':
    diagrama_de_correlação = DiagramaDeCorrelação()
    diagrama_de_correlação.excluindo_colunas()
    diagrama_de_correlação.grafico_com_seaborn()
```
