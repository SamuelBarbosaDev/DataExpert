```python
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
class GráficoDeDispersão():
    """
        Gráfico de dispersão
    """
    def __init__(self):
        self.path = r'../DataFrame/clientes_rename.xlsx'
        self.credit = pd.read_excel(
            self.path
        )
        self.credit_group = self.credit.groupby(
            ["sexo"]
        ).mean().reset_index()
    
    def grafico_com_a_plotly_scatter(self):
        pass

    def situação_dos_clientes_no_mes_setembro(self):
        fig = px.scatter(
            self.credit,
            x='status_pag_set_2005',
            y='valor_pag_set_2005',
            color='estado_civil',
            size='valor_credito',
            title='Sintese Da Situação Dos Clientes'
        )

        fig.show()

        fig.write_html('Gráfico de dispersão.html')
```
