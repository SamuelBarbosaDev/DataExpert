```python
import random
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
class  EstatísticaInferencial():
    def __init__(self):
        path = r'../../data/cost-of-living.csv'
        self.df = pd.read_csv(
            filepath_or_buffer=path,
            sep=',',
            decimal='.',
        )

    def dataframe(self):
        display(self.df)

    def descriptive(self):
        df = self.df.describe()
        display(df)

    def teorema_do_limite_central_prática(self):
        lista = self.df['renda_mensal'].values.tolist()
        means = []

        for i in range(20):
            amostra = random.sample(
                population=lista, 
                k=30
            )
            means.append(np.mean(amostra)) 

        sns.histplot(
            data=means,
            kde = True,
        )
    
    def histplot_renda_mensal(self):
        sns.histplot(
            data=self.df,
            x='renda_mensal',
            kde = True,
        )        

if __name__ == "__main__":
    estatística_inferencial = EstatísticaInferencial()
```

```python
estatística_inferencial.dataframe()
```

```python
estatística_inferencial.descriptive()
```

```python
estatística_inferencial.teorema_do_limite_central_prática()
```

```python
estatística_inferencial.histplot_renda_mensal()
```
