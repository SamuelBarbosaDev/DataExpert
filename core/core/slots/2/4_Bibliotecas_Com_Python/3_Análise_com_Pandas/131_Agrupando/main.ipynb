```python
import pandas as pd
```

```python
class Agrupando():
    """
        Grouping
    """
    def __init__(self):
        self.path = r'../DataFrame/credit_card_clients_2.csv'
        self.credit = pd.read_csv(
            self.path
        )
    
    def agrupando_dados(self):
        credit = self.credit.groupby(
            ['EDUCATION']
        ).mean()
        print(f'\nAgrupando Dados:\n{credit}')
```

```python
if __name__ == '__main__':
    agrupando = Agrupando()
    agrupando.agrupando_dados()
```
